```python
import pandas as pd
from plotnine import *

bio_raw = pd.read_csv ("./data/bio-raw.csv")

bio_raw
```

```python
#| set index as year | sorted index in year order | selected years 2003 to 2013 |#

bio_year = bio_raw.set_index ('year') .sort_index ( ) .loc ['2003' : '2013']

bio_year
```

```python
#| selecting only historical data for accuracy |#

bio_hist = bio_year [bio_year ["scenario"] == "historical"]

bio_hist
```

```python
#| selects for variable method "high intensity agriculture" for consistant, reliable and relivant value ranges |#

bio_hiag = bio_hist [bio_hist ["variable"] == "highintensityag"]

bio_hiag
```

```python
#| removing null value outputs |#

bio_val = bio_hiag [bio_hiag ["value"] >= 0]

bio_val
```

```python
bio_sort = bio_val.drop (columns = ["_id", "lower_uncertainty", "scenario", "upper_uncertainty", "variable"])

bio_sort
```
